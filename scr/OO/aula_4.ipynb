Implemente a seguinte classe:



a) O método palindrome recebe uma String e retorna True se essa String é um palindromo
\*palíndromo é uma palavra que permanece igual quando lida de trás para frente. Exemplo: arara, osso, asa

b) O método contarVogais recebe uma String e retorna a quantidade de vogais contidas na String

Exemplo de entrada

print(StringUtil.palindrome("Ana"))
print(StringUtil.contarVogais("Ana"))

Exemplo de saída:
True
2


```python
class StringUtil:

    @staticmethod
    def palindrome(palavra: str) -> bool:
        if palavra.lower() == palavra[::-1].lower():
            return True
        return False

    @staticmethod
    def contarVogais(palavra: str) -> int:
        vogais: str = "aeiouAEIOU"

        return sum(i in vogais for i in palavra)
```

Implemente a seguinte classe:



a) O construtor dessa classe recebe o nome do aluno

b) O método adicionarNota() recebe um número decimal entre 0 e 10 e adiciona na lista de notas do aluno

c) O método verificarAprovacao() retorna True se a média das notas do Aluno for maior ou igual a 5, caso contrário retorna False

Exemplo de execução:

a1 = Aluno("Glauco")
a1.adicionarNota(5.3)
a1.adicionarNota(0.3)
print(a1.verificarAprovacao()) #deverá retornar False


```python
class Aluno:

    def __init__(self, nome: str) -> None:
        self.nome = nome
        self.__nota = []

    def adicionarNota(self, numero: float) -> None:
        self.__nota.append(numero)

    def verificarAprovacao(self) -> bool:
        if sum(nota for nota in self.__nota) / len(self.__nota) >= 5:
            return True
        else:
            return False
```

```python
a1 = Aluno("Glauco")
a1.adicionarNota(5.3)
a1.adicionarNota(0.3)
print(a1.verificarAprovacao())  # deverá retornar False
```

Implemente a seguinte classe:



a) O construtor dessa classe recebe o nome e a idade

b) O método verificarIdade() retorna True se a pessoa tem 18 anos ou mais, senão retorna False

c) O método qtdeVogais() retorna a quantidade de vogais existentes no nome da Pessoa

d) O método verCompatibilidade() recebe um objeto Pessoa e verifica se esta pessoa é maior de idade, se não retorna False, se sim, então verifica se a quantidade de vogais nos nomes dessas duas pessoas são iguais, se sim, retorna True, senão retorna False

Exemplo de execução:

p1 = Pessoa("Ana", 18)
p2 = Pessoa("Alex", 21)
p3 = Pessoa("Bruno", 8)
print(p1.verCompatibilidade(p3)) #deverá retornar False
print(p1.verCompatibilidade(p2)) #deverá retornar True
print(p2.verCompatibilidade(p3)) #deverá retornar False


```python
class Pessoa:

    def __init__(self, nome: str, idade: int) -> None:
        self.nome = nome
        self.__idade = idade

    def verificarIdade(self) -> bool:
        if self.__idade >= 18:
            return True
        return False

    def qtdeVogais(self) -> int:
        return sum(vogal in self.nome for vogal in 'aeiouAEIOU')

    def verCompatibilidade(self, pessoa: Pessoa) -> bool:
        if pessoa.verificarIdade() and self.qtdeVogais() == pessoa.qtdeVogais():
            return True
        return False
```

```python
p1 = Pessoa("Ana", 18)
p2 = Pessoa("Alex", 21)
p3 = Pessoa("Bruno", 8)
print(p1.verCompatibilidade(p3))  # deverá retornar False
print(p1.verCompatibilidade(p2))  # deverá retornar True
print(p2.verCompatibilidade(p3))  # deverá retornar False
```

Implemente a seguinte classe:



a) O construtor dessa classe deverá receber o valor do atributo andar_maximo. O atributo andar_atual deverá ser inicializado com valor igual a zero.

b) O método subir() deverá aumentar 1 nível no atributo andar_atual (até o limite de andar_maximo). Este método deverá retornar True se o Elevador ainda não atingiu o valor máximo e False se ele estiver no andar_maximo

c) O método descer() deverá reduzir 1 nível no atributo andar_atual (até o limite do andar zero). Este método deverá retornar True se o Elevador ainda não atingiu o andar zero e False se ele estiver no andar zero

d) O método imprimirAndar() deverá retornar um o atributo andar_atual

e) O elevador não poderá descer abaixo do andar 0


```python
class Elevador:

    def __init__(self, andar_maximo: int) -> None:
        self.__andar_maximo = andar_maximo
        self.andar_atual = 0

    def subir(self) -> bool:
        if self.andar_atual < self.__andar_maximo:
            self.andar_atual += 1

            return True
        return False

    def descer(self) -> bool:
        if self.andar_atual > 0:
            self.andar_atual -= 1

            return True
        return False

    def imprimirAndar(self) -> int:
        return self.andar_atual
```

Implemente a seguinte classe:



a) O construtor dessa classe recebe apenas um nmero inteiro que representa o n da conta. O saldo inicial dever ser zero.

b) O mtodo depositar recebe um valor float que dever ser adicionado ao saldo da conta bancaria. Este mtodo retorna a mensagem "deposito realizado com sucesso".

c) O mtodo consultar\_\_saldo retorna o valor do saldo da conta bancaria

d) O mtodo sacar recebe um valor float que reduz o saldo da conta bancria. Se no tiver saldo suficiente dever ser impresso a mensagem "saldo insuficiente", caso contrrio dever ser retornado a mensagem "saldo atualizado"

Exemplo de entrada:
conta1 = ContaBancaria(123)
print(conta1.sacar(10))
print(conta1.consultarSaldo())
print(conta1.depositar(50))
print(conta1.consultarSaldo())
print(conta1.sacar(10))
print(conta1.consultarSaldo())

Exemplo de sada:
saldo insuficiente
0
deposito realizado com sucesso
50
saldo atualizado
40


```python
class ContaBancaria:

    def __init__(self, numero: int) -> None:
        self.numero = numero
        self.__saldo = 0

    def depositar(self, valor: float) -> str:
        self.__saldo += valor
        return "deposito realizado com sucesso"

    def consultarSaldo(self) -> None:
        return self.__saldo

    def sacar(self, valor) -> str:
        if valor > self.__saldo:
            return "saldo insuficiente"
        else:
            self.__saldo -= valor
            return "saldo atualizado"
```

Implemente a seguinte classe:



a) O construtor dessa classe recebe um nome, um valor float e uma quantidade inicial de estoque
b) O mtodo vender recebe um valor inteiro e diminui a quantidade em estoque do produto. Caso no tenha estoque disponvel dever ser retornado a mensagem "estoque indisponvel", caso contrrio retorne a mensagem "estoque atualizado"
c) O mtodo consultarEstoque retorna o valor do produto multiplicado pela quantidade em estoque disponvel do produto

Exemplo de entrada:
arroz = Produto("arroz", 32.10, 50)
print(arroz.consultarEstoque())
print(arroz.vender(60))
print(arroz.vender(10))
print(arroz.consultarEstoque())

Exemplo de saida:
1650.0
estoque indisponvel
estoque atualizado
1284.0


```python
class Produto:

    def __init__(self, nome: str, valor: float, estoque: int) -> None:
        self.nome = nome
        self.__valor = valor
        self.__qtdeEstoque = estoque

    def vender(self, valor: int) -> str:
        if valor > self.__qtdeEstoque:
            return "estoque indisponvel"
        else:
            self.__qtdeEstoque -= valor
            return "estoque atualizado"

    def consultarEstoque(self) -> float:
        return self.__qtdeEstoque * self.__valor
```

```python
arroz = Produto("arroz", 32.10, 50)
print(arroz.consultarEstoque())
print(arroz.vender(60))
print(arroz.vender(10))
print(arroz.consultarEstoque())
```

Implemente a seguinte classe:



a) O construtor dessa classe inicializa uma lista vazia de produtos
b) O mtodo addProduto recebe o nome de um produto e o adiciona na lista de produtos da classe. Se o produto j existir na lista, ento retorne a mensagem: "produto j inserido", caso contrrio retorne a mensagem: "produto inserido"
c) O mtodo listarProdutos imprime todos os produtos inseridos na lista de compras separados por virgula. Se a lista estiver vazia retorne a mensagem: "lista vazia"

lista = ListaCompras()
lista.listarProdutos()
lista.addProduto("arroz")
lista.addProduto("arroz")
lista.addProduto("feijo")
lista.listarProdutos()


```python
class ListaCompras:

    def __init__(self) -> None:
        self.produtos: list[str] = []

    def addProduto(self, nome: str) -> str:
        if nome in self.produtos:
            return "produto j inserido"
        else:
            self.produtos.append(nome)
            return "produto inserido"

    def listarProdutos(self):
        lista = ""
        if self.produtos == []:
            return "lista vazia"
        else:
            for i in range(len(self.produtos)):
                if i == len(self.produtos) - 1:
                    lista += f"{self.produtos[i]}"
                else:
                    lista += f"{self.produtos[i]}, "

        return lista
```

```python
lista = ListaCompras()
print(lista.listarProdutos())
print(lista.addProduto("arroz"))
print(lista.addProduto("arroz"))
print(lista.listarProdutos())
```

Implemente a seguinte classe:



a) O mtodo palindrome recebe uma String e retorna True se essa String  um palindromo
\*palndromo  uma palavra que permanece igual quando lida de trs para frente. Exemplo: arara, osso, asa

b) O mtodo contarVogais recebe uma String e retorna a quantidade de vogais contidas na String

Exemplo de entrada

print(StringUtil.palindrome("Ana"))
print(StringUtil.contarVogais("Ana"))

Exemplo de sada:
True
2


```python
class StringUtil:

    @staticmethod
    def palindrome(palavra: str) -> bool:
        if palavra.lower() == palavra[::-1].lower():
            return True
        return False

    @staticmethod
    def contarVogais(palavra: str) -> int:
        vogais: str = "aeiouAEIOU"

        return sum(i in vogais for i in palavra)
```

```python
print(StringUtil.palindrome("Ana"))
print(StringUtil.contarVogais("Ana"))
```

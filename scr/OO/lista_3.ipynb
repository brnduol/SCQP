Implemente as seguintes classes:



ATENÇÃO: Funcionário não é uma classe abstrata. O seu construtor recebe como parametros o nome e o salario.

a) o método getBonus é o seguinte:
def getBonus(self, x):
return self.salario\*x

b) o método bonifica do Gerente é o seguinte:
def bonifica(self, x):
self.salario = self.salario+super().getBonus(x)

c) o método bonifica do Vendedor é o seguinte:
def bonifica(self, x):
self.salario = self.salario+(self.salario\*x)+self.comissao

Exemplo de entrada:
p1= Gerente("Joao", 1000)
p2 = Vendedor("José", 1000, 50)
p1.bonifica(10)
p2.bonifica(10)
print(p1.salario)
print(p2.salario)

Exemplo de saída:
11000
11050


```python
class Funcionario:

    def __init__(self, nome: str, salario: float) -> None:
        self.nome = nome
        self.salario = salario

    def getBonus(self, x) -> float:
        return self.salario * x


class Gerente(Funcionario):

    def bonifica(self, x) -> None:
        self.salario = self.salario + super().getBonus(x)


class Vendedor(Funcionario):

    def __init__(self, nome: str, salario: float, comissao: float) -> None:
        super().__init__(nome, salario)
        self.comissao = comissao

    def bonifica(self, x):
        self.salario = self.salario + (self.salario * x) + self.comissao
```

```python
p1 = Gerente("Joao", 1000)
p2 = Vendedor("José", 1000, 50)
p1.bonifica(10)
p2.bonifica(10)
print(p1.salario)
print(p2.salario)
```

Complemente o exercício anterior com o método **eq** dentro da classe Funcionário para comparar funcionários. Iremos considerar que 2 funcionários são iguais se eles tiverem o mesmo nome:

Exemplo de entrada:

p1= Gerente("Joao", 1000)

p2 = Vendedor("Joao", 1000, 50)

print(p1 == p2)

Exemplo de saída:
True


```python
class Funcionario:

    def __init__(self, nome: str, salario: float) -> None:
        self.nome = nome
        self.salario = salario

    def __eq__(self, __value: object) -> bool:
        if self.nome == __value.nome:
            return True
        return False

    def getBonus(self, x) -> float:
        return self.salario * x


class Gerente(Funcionario):

    def bonifica(self, x) -> None:
        self.salario = self.salario + super().getBonus(x)


class Vendedor(Funcionario):

    def __init__(self, nome: str, salario: float, comissao: float) -> None:
        super().__init__(nome, salario)
        self.comissao = comissao

    def bonifica(self, x):
        self.salario = self.salario + (self.salario * x) + self.comissao
```

```python
p1 = Gerente("Joao", 1000)

p2 = Vendedor("Joao", 1000, 50)

print(p1 == p2)
```

Crie uma classe chamada Pessoa. Essa classe deverá ter:

a) um atributo publico chamado nome

b) um atributo privado chamado cpf.

c) um atributo estatico chamado qtde que deverá ter valor inicial igual a zero

d) um método estatico chamado incrementaPessoa, cujo objetivo é incrementar o atributo qtde sempre que um novo objeto da classe Pessoa for criado

d) um metodo publico chamado getCPF cujo objetivo é retornar o CPF (int) da pessoa

Exemplo de entrada:

p1 = Pessoa("Joao", 123)

p2 = Pessoa("Jose", 456)

print(Pessoa.qtde)

print(p1.getCPF())

Exemplo de saida:

2
123


```python
class Pessoa:
    qtde = 0

    def __init__(self, nome: str, cpf: int) -> None:
        self.nome = nome
        self.__cpf = cpf
        Pessoa.incrementaPessoa()

    def getCPF(self) -> int:
        return self.__cpf

    @staticmethod
    def incrementaPessoa() -> None:
        Pessoa.qtde += 1
```

```python
p1 = Pessoa("Joao", 123)
p2 = Pessoa("Jose", 456)
print(Pessoa.qtde)
print(p1.getCPF())
```

Implemente as seguintes classes:



a) o método informacoes da classe Livro é o seguinte:

def informacoes(self):
return "este livro chama: " + self.nome + " que tem "+ str(self.qtdePaginas)+ " paginas"

b) o método informacoes da classe Televisao é o seguinte:
def informacoes(self):
return "esta TV chama: " + self.nome + " que tem " + str(self.polegadas)+ " polegadas"

c) o método mostrarCarrinho da classe CarrinhoCompras é o seguinte:
def mostrarCarrinho(self):
for item in self.produtos:
print(item.informacoes())

Exemplo de entrada:

p1 = Televisao("tv21", 21)

p2 = Livro("livro", 10)

c1 = CarrinhoCompras()

c1.adicionar(p1)

c1.adicionar(p2)

c1.mostrarCarrinho()

Exemplo de saida:
esta TV chama: tv21 que tem 21 polegadas
este livro chama: livro que tem 10 paginas


```python
from abc import ABC, abstractmethod


class Produto(ABC):

    def __init__(self, nome: str) -> None:
        super().__init__()
        self.nome = nome

    @abstractmethod
    def informacoes(self):
        ...


class CarrinhoCompras:

    def __init__(self) -> None:
        self.produtos = []

    def adicionar(self, produto: Produto) -> None:
        self.produtos.append(produto)

    def mostrarCarrinho(self):
        for item in self.produtos:
            print(item.informacoes())


class Livro(Produto):

    def __init__(self, nome: str, qtdePaginas: int) -> None:
        super().__init__(nome)
        self.qtdePaginas = qtdePaginas

    def informacoes(self):
        return ("este livro chama: " + self.nome + " que tem " +
                str(self.qtdePaginas) + " paginas")


class Televisao(Produto):

    def __init__(self, nome: str, polegadas: int) -> None:
        super().__init__(nome)
        self.polegadas = polegadas

    def informacoes(self):
        return ("esta TV chama: " + self.nome + " que tem " +
                str(self.polegadas) + " polegadas")
```

```python
p1 = Televisao("tv21", 21)
p2 = Livro("livro", 10)
c1 = CarrinhoCompras()
c1.adicionar(p1)
c1.adicionar(p2)
c1.mostrarCarrinho()
```

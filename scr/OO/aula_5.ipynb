Implemente as seguintes classes



A classe Cliente é uma classe abstrata, ou seja, ela não pode ser instanciada.
A classe Cliente se relaciona com a classe Endereco. Cada Cliente possui apenas 1 endereço.
O método depositar() recebe um valor e incrementa o saldo da conta.
O método sacar() também recebe um valor e decrementa o saldo da conta. Se a conta não tiver dinheiro suficiente a seguinte mensagem deverá ser exibida: "saldo insuficiente"
O método getSaldo() retorna o saldo da conta do cliente.
ATENÇÃO: o atributo saldo é um atributo protegido

Exemplo de execução:

e1 = Endereco("NSWT", 123, "Brasilia")
e2 = Endereco("SQS", 77, "Gama")
c1 = PFisica(123, "Glauco", e1)
c2 = PJuridica(222, "FGA", e2)
print(c1.endereco.cidade)
print(c2.endereco.cidade)


```python
from abc import ABC, abstractmethod


class Endereco:

    def __init__(self, rua: str, numero: str, cidade: str) -> None:
        self.rua = rua
        self.numero = numero
        self.cidade = cidade


class Cliente(ABC):

    def __init__(self, nome: str, endereco: Endereco) -> None:
        super().__init__()
        self.nome = nome
        self._saldo = 0
        self.endereco = endereco

    def depositar(self, valor: float) -> None:
        self._saldo += valor

    def sacar(self, valor: float) -> None:
        if self._saldo >= valor:
            self._saldo -= valor
            return
        return 'saldo insuficiente'

    def getSaldo(self) -> float:
        return self._saldo


class PFisica(Cliente):

    def __init__(self, nome: str, cpf: int, endereco: Endereco) -> None:
        super().__init__(nome, endereco)
        self.cpf = cpf


class PJuridica(Cliente):

    def __init__(self, nome: str, cnpj: int, endereco: Endereco) -> None:
        super().__init__(nome, endereco)
        self.cnpj = cnpj
```

```python
e1 = Endereco("NSWT", 123, "Brasilia")
e2 = Endereco("SQS", 77, "Gama")
c1 = PFisica(123, "Glauco", e1)
c2 = PJuridica(222, "FGA", e2)
c1.depositar(100)
c2.depositar(100)
print(c1.sacar(200))
print(c2.sacar(200))
```

A classe Agencia tem um atributo estático chamado qtdeClientes. Esse atributo deverá ser incrementado toda vez que um novo cliente for adicionado à Agencia.
O método addCliente recebe um objeto do tipo Cliente e adiciona na lista de clientes da Agencia.
O método mostrarSaldo() retorna a soma dos saldos de TODOS os clientes da Agencia


```python
from abc import ABC, abstractmethod


class Endereco:

    def __init__(self, rua: str, numero: str, cidade: str) -> None:
        self.rua = rua
        self.numero = numero
        self.cidade = cidade


class Cliente(ABC):

    def __init__(self, nome: str, endereco: Endereco) -> None:
        super().__init__()
        self.nome = nome
        self._saldo = 0
        self.endereco = endereco

    def depositar(self, valor: float) -> None:
        self._saldo += valor

    def sacar(self, valor: float) -> None:
        if self._saldo >= valor:
            self._saldo -= valor
            return
        return 'saldo insuficiente'

    def getSaldo(self) -> float:
        return self._saldo


class Agencia:
    qtdeClientes: int = 0

    def __init__(self, codigo: int) -> None:
        self.codigo = codigo
        self.cliente = []

    def addCliente(self, novoCliente: Cliente) -> None:
        self.cliente.append(novoCliente)
        self.aumentarClientes()

    def mostrarSaldo(self) -> float:
        return sum(i.getSaldo() for i in self.cliente)

    @classmethod
    def aumentarClientes(cls) -> None:
        cls.qtdeClientes += 1


class PFisica(Cliente):

    def __init__(self, cpf: int, nome: str, endereco: Endereco) -> None:
        super().__init__(nome, endereco)
        self.cpf = cpf


class PJuridica(Cliente):

    def __init__(self, cnpj: int, nome: str, endereco: Endereco) -> None:
        super().__init__(nome, endereco)
        self.cnpj = cnpj
```

```python
e1 = Endereco("NSWT", 123, "Brasilia")
e2 = Endereco("SQS", 77, "Gama")
c1 = PFisica(123, "Glauco", e1)
c2 = PJuridica(222, "FGA", e2)
c1.depositar(100)
c2.depositar(80)
a = Agencia(222)
a.addCliente(c1)
a.addCliente(c2)
print(a.cliente[0].nome)
print(a.cliente[1].nome)
```

```python
from abc import ABC, abstractmethod


class Endereco:

    def __init__(self, rua: str, numero: str, cidade: str) -> None:
        self.rua = rua
        self.numero = numero
        self.cidade = cidade


class Cliente(ABC):

    def __init__(self, nome: str, endereco: Endereco) -> None:
        super().__init__()
        self.nome = nome
        self._saldo = 0
        self.endereco = endereco

    def depositar(self, valor: float) -> None:
        self._saldo += valor

    def sacar(self, valor: float) -> None:
        if self._saldo >= valor:
            self._saldo -= valor
            return
        return 'saldo insuficiente'

    def getSaldo(self) -> float:
        return self._saldo


class Agencia:
    qtdeClientes: int = 0

    def __init__(self, codigo: int) -> None:
        self.codigo = codigo
        self.cliente = []

    def addCliente(self, novoCliente: Cliente) -> None:
        self.cliente.append(novoCliente)
        self.aumentarClientes()

    def mostrarSaldo(self) -> float:
        return sum(i.getSaldo() for i in self.cliente)

    @classmethod
    def aumentarClientes(cls) -> None:
        cls.qtdeClientes += 1


class Banco:

    def __init__(self, codigo):
        self.codigo = codigo
        self.agencia = {}

    def addAgencia(self, int) -> None:
        self.agencia[int] = Agencia(int)

    def mostrarSaldo(self) -> float:
        return sum(self.agencia[i].mostrarSaldo() for i in self.agencia)


class PFisica(Cliente):

    def __init__(self, cpf: int, nome: str, endereco: Endereco) -> None:
        super().__init__(nome, endereco)
        self.cpf = cpf


class PJuridica(Cliente):

    def __init__(self, cnpj: int, nome: str, endereco: Endereco) -> None:
        super().__init__(nome, endereco)
        self.cnpj = cnpj
```

```python
e1 = Endereco("NSWT", 123, "Brasilia")
e2 = Endereco("SQS", 77, "Gama")
c1 = PFisica(123, "Glauco", e1)
c2 = PJuridica(222, "FGA", e2)
c1.depositar(100)
c2.depositar(80)

b = Banco(777)
b.addAgencia(101)
b.addAgencia(102)

b.agencia[101].addCliente(c1)
b.agencia[102].addCliente(c2)


print(Agencia.qtdeClientes)
print(b.mostrarSaldo())
```
